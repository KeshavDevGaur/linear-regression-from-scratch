```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([500, 1000, 1500, 2000, 2500], dtype='float')
Y = np.array([50, 100, 150, 200, 250], dtype='float')
```

```python
X_mean = np.mean(X)
X_std = np.std(X)
X = (X - X_mean) / X_std
```

```python

m = 0   
b = 0   

learning_rate = 0.01  
epochs = 100          

n = len(X)
```

```python

costs = []


for i in range(epochs):

  
    Y_pred = m * X + b
 
    cost = (1/n) * np.sum((Y - Y_pred)**2)
    costs.append(cost) 
    dm = (-2/n) * np.sum(X * (Y - Y_pred)) 
    db = (-2/n) * np.sum(Y - Y_pred)        

    m = m - learning_rate * dm
    b = b - learning_rate * db
```

```python
print("Slope (m):", m)        
print("Intercept (b):", b)    
```

```python
new_x = 3000
new_x_norm = (new_x - X_mean) / X_std
predicted_price = m * new_x_norm + b

print("Predicted Price for 3000 sq ft:", predicted_price)
```

```python
plt.scatter(X, Y)
plt.plot(X, m * X + b)

plt.xlabel("Normalized Size")
plt.ylabel("Price")
plt.title("Linear Regression Fit")
plt.show()
```

```python
plt.plot(costs)

plt.xlabel("Iterations")
plt.ylabel("Cost (MSE)")
plt.title("Cost Reduction Over Time")
plt.show()
```
